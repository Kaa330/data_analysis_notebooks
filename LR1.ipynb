### 7. Выявление и устранение пропусков

```python
print(data.isna().sum())
```

Данной командой мы выяснили, в столбцах Total_Bilirubin, Alkaline_Phosphotase, Albumin_and_Globulin_Ratio есть пропуски. Как мы видими, мне повезло и данные пропуски присутствуют именно в конце датасета, что легко наглядно отобразить. Я решил, что устранить пропуски будет рационально удалить строки с пустыми значениями полностью, так как если бы этот датасет использовался как база данных для медицинского исследования, то заменив пустые значения на нули или среднии значения, то это сильно бы могло повлиять на результаты. В таких сферах важна точность и достоверность данных.

```python
data.isna()
```

```python
data = data.dropna(subset=['Total_Bilirubin', 'Alkaline_Phosphotase', 'Albumin_and_Globulin_Ratio'])
```

Повторим команду для проверки конца датасета

### 9. Проверка типов данных.

```python
data.info()
```

Столбцу Aspartate_Aminotransferase стоит назначить числовой тип данных.
Мы используем функцию to_numeric с параметром errors = 'coerce', а значит не корректные данные не выдадут ошибку, а просто заменятся на NaN. Поэтому повторим удаление строк с пустыми значениями с помощью метода .dropna.
А столбец Group преобразовать в категориальныю переменную.

```python
data['Aspartate_Aminotransferase'] = pd.to_numeric(data['Aspartate_Aminotransferase'], errors='coerce')
```

```python
data.dtypes
```

```python
data = data.dropna(subset=['Aspartate_Aminotransferase'])
```

```python
data.isna()
```

Видим, что в наших данных удалилась еще одна строка, а тип Aspartate_Aminotransferase изменился на float64.

Cтолбец Group преобразовать в категориальныю переменную. Перед этим преобразуем эту переменную в 89 строке к значению - 1.

```python
data.head(85)
```

```python
data.drop(labels = 89, axis=0, inplace=True)
data['Group'] = data['Group'].astype('category')
```

Так же удалим строки с некоректными значениями из столбца Gender

```python
data.drop(data[data['Gender'] == 'Mal'].index, inplace=True)
```

```python
data.dtypes
```

```python
data.head(85)
```

Мы удалили строчки с некотректными данными и сделали переменную Grope категориальной с категориями 1 и 2.

### 10. Создание сводных таблиц.

```python
ptable = pd.pivot_table(data, index = ['Group'], values = ['Albumin', 'Alkaline_Phosphotase', 'Total_Bilirubin'])
```

```python
ptable.head(30)
```

Сооздал сводную таблицу по столбцу "Group" где данны среднии показатели альбумина, щелочной фосфатазы и общего билирубина для группы больных и здоровых. 

### 11. Сделайте еще две группировки с помощью group_by или pivot_table. <a id='11'></a>

```python
gtable = data.groupby(["Group", "Gender"])["Gender"].count()
```

```python
gtable.head(15)
```

При помощи данной группиировки наглядно можно определить сколько представителей конкретного пола находится в числе заболевших и здоровых.

```python
ptable2 = pd.pivot_table(data,
                         index = ['Group'],
                         values = ['Albumin'],
                         columns=['Gender'],
                         aggfunc=[np.mean, len]
                        )

```

```python
ptable2.head(16)
```

Сооздал сводную таблицу по столбцу "Group" где данны среднии показатели альбумина для женщин и мужчин в разных группах, а так же количество женщин и мужчин в разных группах.
